```python
import pandas as pd
```

```python
osoby = pd.read_excel('./data/osoby.xlsx')
```

```python
osoby.concatenatedWorkingPositionOrganizations.value_counts()
```

```python
osoby = osoby[osoby.active == True]
```

```python
osoby.columns
```

```python
def gender(name):
    if name.endswith('á'):
        return 'zena'
    else:
        return 'muz'
```

```python
osoby['gender'] = osoby.lastName.apply(lambda x: gender(x))
```

```python
os = osoby.to_dict(orient='index')
```

```python
osoby[osoby.concatenatedWorkingPositionOrganizations.str.contains('soud ')].concatenatedWorkingPositions.value_counts()
```

```python
out = {}
for o in os:
    g = os[o]['gender']
    org = os[o]['concatenatedWorkingPositionOrganizations']
    
    if org not in out:
        out[org] = {'muzi': 0, 'zeny': 0}
    
    if g == 'zena':
        out[org]['zeny'] += 1
    else:
        out[org]['muzi'] += 1
```

```python
for o in out:
    out[o]['pomer'] = out[o]['zeny'] / (out[o]['zeny'] + out[o]['muzi'])
```

```python
dta = pd.DataFrame.from_dict(out, orient='index')
```

```python
dta['sum'] = dta.apply(lambda row: row['zeny'] + row['muzi'], axis=1)
```

```python
dta.sort_values(by='sum', ascending=False)
```
